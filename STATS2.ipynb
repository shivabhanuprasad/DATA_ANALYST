```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define function FIRST
def detect_outliers(data, threshold=3):
    mean = np.mean(data)
    std = np.std(data)

    z_scores = [(x - mean) / std for x in data]

    outliers = []
    for i, z in enumerate(z_scores):
        if abs(z) > threshold:
            outliers.append(i)   # return value

    return outliers
```

```python
# define our dataset
data=[11,10,12,14,15,16,15,16,12,13,14,17,18,19,12,11,10,80,100,110]
```

```python
outliers=detect_outliers(data)
```

```python
print("Outliers:",outliers)
```

# IQR
1. Sort the data
2. Calculate Q1 and Q3
3. IQR(Q3-Q1)
4. Find the Lower Fence(q1-1.5(IQR))
5. Find the upper Fence ((q1+1.5(IQR)))  

```python
data=[10,11,12,12,13,15,15,13,15,13,13,11,10,11,14,12,14,108,102]
```

```python
data=sorted(data)
data
```

```python
import numpy as np
q1,q3= np.percentile(data,[25,75])
```

```python
print(q1,q3)
```

```python
iqr=q3-q1
print(iqr)
```

```python
# Find the Lower Fence and Higher Fence
lower_fence= q1-(1.5*iqr)
higher_fence= q3+(1.5*iqr)
print(lower_fence,higher_fence)
```

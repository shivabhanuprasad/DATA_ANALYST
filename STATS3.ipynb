```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
# enter IQ levels of 20 patients
data=[88,92,94,94,96,97,97,97,99,99,105,109,109,110,112,112,113,114,115]
ztest(data,value=100)
```

```python
# consider another example
# ages of the college students(population)
#  1 class students mean if all the ages
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
school_ages
```

```python
classA_ages
```

```python
classA_ages.mean()
```

```python
_,p_value=ttest_1samp(classA_ages,popmean=school_ages.mean())
```

```python
school_ages.mean()
```

```python
if p_value<=0.05:
    print("Accept H0")
else:
    print("Reject H0")
```
